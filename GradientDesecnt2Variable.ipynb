```python
from sklearn.datasets import make_regression
X,y=make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=5)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=13)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
y_pred=lr.predict(X_test)
```

```python
plt.scatter(X,y)
plt.plot(X_test,y_pred,color='red')
```

```python
print(lr.intercept_,lr.coef_)
```

```python
import numpy as np
```

```python
class GDRegresion:
    def __init__(self,epochs,lr):
        self.b=100
        self.m=100
        self.epochs=epochs
        self.lr=lr
    def fit(self,X,y):
        for i in range(self.epochs):
            lose_slop_b=-2*np.sum(y-self.m*X.ravel()-self.b)
            self.b=self.b-(self.lr*lose_slop_b)
            lose_slop_m=-2*np.sum((y-self.m*X.ravel()-self.b)*X.ravel())
            self.m=self.m-(self.lr*lose_slop_m)
        print(self.b,self.m)
```

```python
gr=GDRegresion(20,0.01)
```

```python
gr.fit(X_train,y_train)
```
